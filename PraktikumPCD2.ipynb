```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import matplotlib as mpl
```

```python
img = cv2.imread('/content/Cat.jpg')
```

```python
photo = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img)
plt.show()

kernel = np.ones((5,5),np.float32)/25
print(kernel)

photo_filter = cv2.filter2D(img,-1,kernel)

plt.imshow(photo_filter)
```

```python
plt.rcParams["figure.figsize"] = (15,15)

plt.subplot(121),plt.imshow(photo),plt.title('Original')
plt.xticks([]), plt.yticks([])

plt.subplot(122),plt.imshow(photo_filter),plt.title('Averaging')
plt.xticks([]), plt.yticks([])

plt.show()
```

```python
photo_blur = cv2.blur(img,(5,5))

plt.imshow(photo_blur)
```

```python
kernel = np.matrix([
          [1, 1, 1],
          [1, 2, 1],
          [1, 1, 1]         
          ])/25
print(kernel)

photo_filter = cv2.filter2D(img,-1,kernel)

plt.imshow(photo_filter)
```
